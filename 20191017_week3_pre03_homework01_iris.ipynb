## 【問題1】データの取得

```python
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
#print(iris)

X = pd.DataFrame(iris.data,columns=iris.feature_names)
print(X.shape)

y = pd.DataFrame(iris.target,columns={'Species'})
print(y.shape)
print(iris.target_names)
```

## 【問題2】データの結合

```python
df = pd.concat([X, y], axis=1)
print(df.shape)
```

## 【問題6】図の作成

```python
#%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
#ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）

#ラベルごとにデータを分割   df[df['Species'].isin(['i'])]

for i in range(len(iris.target_names)):
    df_i = df[df['Species'].isin(['i'])]
    
    plt.style.use('default')
    sns.set()
    sns.set_style('whitegrid')
    sns.set_palette('Set1')
    
    
    x = iris.feature_names
    y = np.array([len(df.iloc[:,0]), len(df.iloc[:,1]),
                  len(df.iloc[:,2]), len(df.iloc[:,3])])

    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.pie(y, labels=x, autopct="%1.1f%%")
    ax.set_title(iris.target_names[i])



plt.show()
```

```python
#特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する

# タイトルが特徴量、x軸がラベル、y軸が特徴量データ。

# まず特徴量ごとにデータを抜き出す #df_i = df.iloc[:,i]
#抜き出したデータのラベルごとの統計量を出す(必要なら)　＝　y軸の特徴量データ
#df_j = df[df['Species'].isin([j])]
#df_i = df_j.iloc[:,i]
#描画


for i in range(len(iris.feature_names)):    
    df_i = df.iloc[:,i] #特徴量のデータを抽出
        
    datas_each = (df_i[df['Species'].isin([0])],
                  df_i[df['Species'].isin([1])],
                  df_i[df['Species'].isin([2])]) 
    #特徴量×ラベルごとの全データを抽出
    
    # 箱ひげ図
    fig, ax = plt.subplots()

    bp = ax.boxplot(datas_each)
    ax.set_xticklabels(iris.target_names)
        
    # タイトル名  
    plt.title(iris.feature_names[i])
    
    plt.xlabel('labels')
    plt.ylabel('points')
        
    plt.show()
```

```python
#特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する
for i in range(len(iris.feature_names)):    
    df_i = df.iloc[:,i] #特徴量のデータを抽出
        
    datas_each = (df_i[df['Species'].isin([0])],
                  df_i[df['Species'].isin([1])],
                  df_i[df['Species'].isin([2])]) 
    #特徴量×ラベルごとの全データを抽出
    
    # バイオリン図
    sns.set(style='whitegrid')
    ax = sns.violinplot(data=datas_each)
        
    # タイトル名  
    plt.title(iris.feature_names[i])
    
    plt.show()
```

```python
#４つの特徴量に関する相関係数行列を作る
corr_mat = X.corr(method='pearson')
print(corr_mat)
```

```python
#相関係数行列のヒートマップを作る
sns.heatmap(corr_mat,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に値を表示
            fmt='.1f',
            xticklabels=corr_mat.columns.values,
            yticklabels=corr_mat.columns.values
           )
```
