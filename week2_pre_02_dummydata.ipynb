# 【問題1】乱数の作成

```python
import numpy as np

np.random.seed(0)

mu = [-3,0]
sigma = [[1.0,0.8],[0.8,1.0]]

#二次元正規乱数を500個生成
values = np.random.multivariate_normal(mu,sigma,500)

print(values.shape)
```

# 【問題2】散布図による可視化

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.scatter(values[:,0],values[:,1],marker="*")
```

```python
sns.jointplot(values[:,0], values[:,1])
```

# 【問題3】ヒストグラムによる可視化

```python
plt.hist(values[:,0])
plt.xlim(-6,3)
plt.show()

plt.hist(values[:,1])
plt.xlim(-6,3)
plt.show()
```

```python
fig, ax=plt.subplots(2,1)

ax[0].hist(values[:,0],bins=100,range=(-6,3))
ax[0].set_title("hist01")
ax[0].set_xlabel("x1")
ax[0].set_ylabel("frequency")

ax[1].hist(values[:,1],bins=100,range=(-6,3))
ax[1].set_title("hist02")
ax[1].set_xlabel("x2")
ax[1].set_ylabel("frequency")
```

# 【問題4】データの追加

```python
np.random.seed(0)

mus = [0,-3]
sigmas = [[1.0,0.8],[0.8,1.0]]

#二次元正規乱数を500個生成
values_s = np.random.multivariate_normal(mus,sigmas,500)

print(values_s.shape)
```

```python
#問題1、4それぞれのデータをひとつの散布図に。問題1のものは0、問題４のものは1を表示
plt.plot(values[:,0],values[:,1],'o', label='0')
plt.plot(values_s[:,0],values_s[:,1],'*', label='1')

plt.legend(loc='upper right')
```

# 【問題5】データの結合
>問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。

```python
combi = np.vstack((values,values_s))
print(combi)
```

# 【問題6】ラベル付け
>問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。
0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。

```python
zeros = np.reshape(np.zeros(500),(500,1))
#print(zeros)

ones = np.reshape(np.ones(500),(500,1))
#print(ones)

#ラベルを結合
labels = np.vstack((zeros,ones))

labeled = np.hstack((combi,labels))
print(labeled)
```
